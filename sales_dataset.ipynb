```python
import pandas as pd

# -----------------------------
# Create Sales Dataset
# -----------------------------
data = {
    "Order_ID": [1001, 1002, 1003, 1004, 1005],
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Printer"],
    "Category": ["Electronics", "Accessories", "Accessories", "Electronics", "Electronics"],
    "Quantity": [2, 5, 3, 1, 4],
    "Price": [50000, 500, 1500, 12000, 8000]
}

sales_df = pd.DataFrame(data)

# Add Total Sales column
sales_df["Total_Sales"] = sales_df["Quantity"] * sales_df["Price"]

# -----------------------------
# Analysis
# -----------------------------
total_revenue = sales_df["Total_Sales"].sum()
category_sales = sales_df.groupby("Category")["Total_Sales"].sum()
best_product = sales_df.loc[sales_df["Quantity"].idxmax()]

# -----------------------------
# Output
# -----------------------------
print("SALES DATASET\n")
print(sales_df)

print("\nTotal Revenue:", total_revenue)

print("\nSales by Category:\n")
print(category_sales)

print("\nBest Selling Product (by Quantity):\n")
print(best_product)

# -----------------------------
# Save to CSV
# -----------------------------
sales_df.to_csv("sales_data.csv", index=False)
```
